```python
def p2f(x):
    return float(x.strip('%'))/100

df['Percent of Students Chronically Absent']=df['Percent of Students Chronically Absent'].astype(str).apply(p2f)
df['Rigorous Instruction %'] = df['Rigorous Instruction %'].astype(str).apply(p2f)
df['Collaborative Teachers %'] = df['Collaborative Teachers %'].astype(str).apply(p2f)
df['Supportive Environment %'] = df['Supportive Environment %'].astype(str).apply(p2f)
df['Effective School Leadership %'] = df['Effective School Leadership %'].astype(str).apply(p2f)
df['Strong Family-Community Ties %'] = df['Strong Family-Community Ties %'].astype(str).apply(p2f)
df['Trust %'] = df['Trust %'].astype(str).apply(p2f)
df['Student Attendance Rate'] = df['Student Attendance Rate'].astype(str).apply(p2f)
df['Percent Asian'] = df['Percent Asian'].astype(str).apply(p2f)
df['Percent Black'] = df['Percent Black'].astype(str).apply(p2f)
df['Percent Hispanic'] = df['Percent Hispanic'].astype(str).apply(p2f)
df['Percent White'] = df['Percent White'].astype(str).apply(p2f)
```

```python
#4 Plot
sns.set(style='ticks')
sns.boxplot(data=df['Percent Asian'], palette='PRGn')
sns.despine(offset=10, trim=True)


```

```python
#Two continuous variables 3 different ways: Black percentages vs. Hispanic
#1
grid = sns.JointGrid(df['Percent Hispanic'], df['Percent Black'], space=0, size=6, ratio=50)
grid.plot_joint(plt.scatter, color='g')
grid.plot_marginals(sns.rugplot, height=1, color='g')
```

```python
#One continuous vs. one categorical variable. 6 graphs
#Categorical variable will be community school
community_school = df[df["Community School?"]=='Yes']
no_community = df[df['Community School?']=='No']
```

```python
#1&2
plt.subplot(1,2,1)
sns.set(style='ticks')
sns.boxplot(data=community_school['Percent Hispanic'], palette='PRGn')
sns.despine(offset=10, trim=True)

plt.subplot(1,2,2)
sns.set(style='ticks')
sns.boxplot(no_community['Percent Hispanic'], palette='PRGn')
sns.despine(offset=10, trim=True)
```

```python
#3&4
plt.subplot(1,2,1)
sns.distplot(no_community['Percent White'], kde=False, color='b')

plt.subplot(1,2,2)
sns.distplot(community_school['Percent White'], kde=False, color='g')
```
