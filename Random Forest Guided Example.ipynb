```python
#Changing Loan Status to Categorical Values

def cat_to_num(x):
    if x=='Default':
        return 6
    if x=='Late (31-120 days)':
        return 5
    if x=='Late (16-30 days)':
        return 4
    if x=='In Grace Period':
        return 3
    if x=='Current':
        return 2
    if x=='Fully Paid':
        return 1
    if x=='Charged off':
        return 0
```

```python
y2015['loan_status_cat'] = y2015['loan_status'].apply(cat_to_num)
```
